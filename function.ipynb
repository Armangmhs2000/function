```python
def mulp1():
    x=int(input('x= '))
    y=int(input('y= '))
    multiplication=x*y
    print('multiplication= ',multiplication)
    return multiplication
def mulp2():
    z=int(input('z = '))
    total=mulp1()
    total2=total*z
    print ('total2 = ',total2)  
#how we can use multiple functions simultaneously
```
